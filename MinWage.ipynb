```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns

df = pd.read_csv('Minimum Wage Data.csv', encoding='latin')
df.to_csv('minwage.csv', encoding='utf-8')
```

```python
df = pd.read_csv('minwage.csv')
df = df.set_index('Year')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df
```

```python
grouped_by_state=df.groupby('State')
#grouped_by_state[['State.Minimum.Wage.2020.Dollars']].apply(print)
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby('State'):
    if act_min_wage.empty:
        act_min_wage = group[['State.Minimum.Wage.2020.Dollars']].rename(columns={'State.Minimum.Wage.2020.Dollars':name})
    else:
        act_min_wage = act_min_wage.join(group[['State.Minimum.Wage.2020.Dollars']].rename(columns={'State.Minimum.Wage.2020.Dollars':name}))
act_min_wage
```

```python
act_min_wage.describe()
```

```python
act_min_wage.corr()
corr_min_wage = act_min_wage.corr().fillna(0)
corr_min_wage
```

```python
labels = [c[:2] for c in corr_min_wage.columns]  # get abbv state names.

fig = plot.figure(figsize=(20,15))  # figure so we can add axis
ax = fig.add_subplot(111)  # define axis, so we can modify
ax.matshow(corr_min_wage, cmap=plot.cm.RdYlGn)  # display the matrix
ax.set_xticks(np.arange(len(labels)))  # show them all!
ax.set_yticks(np.arange(len(labels)))  # show them all!
ax.set_xticklabels(labels)  # set to be the abbv (vs useless #)
ax.set_yticklabels(labels)  # set to be the abbv (vs useless #)

plot.show()
```
